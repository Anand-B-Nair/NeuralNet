```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
from google.colab import files
```

Data Collection and Analysis

```python
store_data = pd.read_csv('/content/train.csv')
```

```python
store_data.shape
```

```python
store_data.info()
```

```python
store_data=store_data.dropna(how='all')
store_data.info()
store_data.shape
```

```python
print(store_data['state'].unique())
print(store_data['product'].unique())
print(store_data['store'].unique())
```

Label Encoder

```python
encoder = LabelEncoder()
store_data['date'] = encoder.fit_transform(store_data['date'])
store_data['state'] = encoder.fit_transform(store_data['state'])
store_data['store'] = encoder.fit_transform(store_data['store'])
store_data['product'] = encoder.fit_transform(store_data['product'])
store_data.head(20)
```

Training & Testing

```python
X = store_data.drop(columns='num_sold',axis=1)
Y = store_data['num_sold']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

XGBREGRESSION


```python
regr = XGBRegressor()
regr.fit(X_train, Y_train)
train_pred = regr.predict(X_train)
r2_train = metrics.r2_score(Y_train, train_pred)
print('R2 value = ', r2_train)
```

```python
test_pred = regr.predict(X_test)
r2_test = metrics.r2_score(Y_test, test_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(Y_test, test_pred)
mse = mean_squared_error(Y_test, test_pred)
r2 = r2_score(Y_test, test_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

Hyperparameter Tuning for XGB

```python
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb

param_dist = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.3],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 0.1, 0.3],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [0, 0.1, 0.5]
}

xgb_model = xgb.XGBRegressor()

random_search = RandomizedSearchCV(estimator=xgb_model, param_distributions=param_dist, n_iter=100, cv=3, scoring='neg_mean_squared_error', verbose=2, random_state=2)
random_search.fit(X_train, Y_train)

best_params = random_search.best_params_
print("Best Hyperparameters:", best_params)

best_model = random_search.best_estimator_

predictions = best_model.predict(X_test)

mae = mean_absolute_error(Y_test, predictions)
mse = mean_squared_error(Y_test, predictions)
r2 = r2_score(Y_test, predictions)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
from sklearn.model_selection import KFold, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error, r2_score

xgb_model = xgb.XGBRegressor()

scorer = make_scorer(mean_squared_error, greater_is_better=False)

param_dist = {
    'regressor__n_estimators': np.arange(100, 501, 50),
    'regressor__max_depth': np.arange(3, 10),
    'regressor__learning_rate': [0.1, 0.3, 0.5],
    'regressor__min_child_weight': np.arange(1, 6),
    'regressor__gamma': [0, 0.1, 0.3],
    'regressor__subsample': [0.8, 0.9, 1.0],
    'regressor__colsample_bytree': [0.8, 0.9, 1.0],
    'regressor__reg_alpha': [0, 0.1, 0.5],
    'regressor__reg_lambda': [0, 0.1, 0.5]
}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

pipeline = Pipeline([
    ('regressor', xgb_model)
])

random_search = RandomizedSearchCV(estimator=pipeline, param_distributions=param_dist, n_iter=100, cv=kf, scoring=scorer, verbose=2, n_jobs=-1, random_state=42)
random_search.fit(X_train, Y_train)

best_params = random_search.best_params_
print("Best Hyperparameters:", best_params)

best_model = random_search.best_estimator_

predictions = best_model.predict(X_test)

mae = mean_absolute_error(Y_test, predictions)
mse = mean_squared_error(Y_test, predictions)
r2 = r2_score(Y_test, predictions)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```
